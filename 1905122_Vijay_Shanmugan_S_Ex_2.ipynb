```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/VijayShanmugan/Datasets/main/Tokyo%20Medals%202021.csv')
df[:30]
```

```python
plt.plot(df['Rank By Total'], df['Total'], color='blue', marker='o')
plt.title('Relationship between Rank and Medals', fontsize=16)
plt.xlabel('Rank', fontsize=14)
plt.ylabel('Total medals', fontsize=14)
plt.grid(True)
plt.show()
```

```python
sns.regplot(x="Rank By Total", y="Total", data=df);
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(16,6))

x=df["Rank By Total"][:20]
plt.title('Gold medals Vs Silver medals')

plt.xlabel('Rank')
plt.ylabel('Medals')

plt.plot(x,df["Gold Medal"][:20],"+--",c="r",label="Gold")
plt.plot(x,df["Silver Medal"][:20],"*--",c="b",label="Silver")

plt.legend()
plt.show()
```
