```python
goldmedal_one=np.asarray([39,28,27,22,20,17,10,10,10,10,7,7,7,7,4,4,4,4])
total_one=np.asarray([113,88,58,65,71,46,36,33,37,40,24,21,20,15,20,20,14,11])
bronzemedal_one=np.asarray([33,13,17,22,23,22,14,11,10,8,9,17,9,5,3,2,6,0])
countryposition_one = [low, high, medium, medium, high, low]*3

goldmedal_two=np.asarray([4,2,2,2,1,1,1,1,0,0,0,0])
total_two=np.asarray([10,8,9,17,9,13,11,8,7,9,7,6])
bronzemedal_two=np.asarray([5,15,16,22,15,10,2,6,26,8,14,20])
countryposition_two = [low, high, medium, medium, high, low]*2

plt.scatter(
    x=goldmedal_one,
    y=total_one,
    s=bronzemedal_one * 10,
    c=countryposition_one,
    alpha=0.5
)
plt.scatter(
    x=goldmedal_two,
    y=total_two,
    s=bronzemedal_two * 10,
    c=countryposition_two,
    marker="d",
    alpha=0.5
)

plt.title("Contribution of Gold medals in Total medals")
plt.legend(["Gold medals", "Total medals"])
plt.xlabel("Gold medals")
plt.ylabel("Total medals")

plt.show()
```

```python
plt.scatter(
    x=goldmedal_one,
    y=total_one,
    s=bronzemedal_one * 10,
    c=countryposition_one,
    cmap="jet",
    alpha=0.5
)
plt.scatter(
    x=goldmedal_two,
    y=total_two,
    s=bronzemedal_two * 10,
    c=countryposition_two,
    marker="d",
    cmap="jet",
    alpha=0.5
)

plt.title("Contribution of Gold medals in Total medals")
plt.legend(["Gold medals", "Total medals"])
plt.xlabel("Gold medals")
plt.ylabel("Total medals")

plt.colorbar()
plt.show()
```
